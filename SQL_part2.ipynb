```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
db = sql.connect('electricity.db')
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type = 'table'")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(elect)")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elect LIMIT 15",db,index_col = 'index')
```

```python
# selecting distinct elements from the sql file
query = "SELECT DISTINCT country FROM elect"
country = pd.read_sql(query,db)
```

```python
'Uzbekistan' in country['country'].values
```

```python
query = "SELECT * FROM elect WHERE country = 'Uzbekistan'  AND year >2010 "
pd.read_sql(query, db).head()
```

```python
# Between method
query = "SELECT * FROM elect WHERE country = 'Uzbekistan' AND year BETWEEN 2010 AND 2020"
pd.read_sql(query,db,index_col = 'index')
```

```python
#in method
query = "SELECT * FROM elect WHERE country = 'Uzbekistan' AND year IN (2010,2011,2019,2018)"
pd.read_sql(query,db,index_col = 'index')
```

```python
# like method
# %var Anvar
# _ar - car,tar
# _ol_- xola
# %ol% - bola
#example

query = "SELECT * FROM elect WHERE country LIKE '%stan' and year IN (2019,2010)"
pd.read_sql(query,db,index_col = 'index')
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' and year IN (2019,2018)"
pd.read_sql(query,db,index_col = 'index')
```
