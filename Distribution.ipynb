```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv')
df = df[df.year >= 2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
df.shape
```

# Histogram(histplot)

```python
plt.figure(figsize = (10,5))
sns.histplot(data = df, x = 'price')
plt.show()
```

# Displot()

```python
sns.displot(data = df, x = 'price',height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price',hue = 'transmission',height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price',hue = 'transmission',multiple = 'stack',height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price',col = 'transmission',height = 5, aspect = 2) #ploting 3 different graphs for each type of transmission
plt.show()
```

#KDE

```python
sns.displot(data = df, x = 'price',kind = 'kde',height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price',kde = True,height = 5, aspect = 2)
plt.show()
```

```python
# kdeplot
plt.figure(figsize = (10,5))
sns.kdeplot(data = df, x = 'price', shade = True)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.kdeplot(data = df, x = 'price',hue = 'fuelType', shade = True)
plt.show()
```

```python
sns.displot(data = df, x = 'price', hue = 'fuelType', kind = 'kde', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.show()
```

```python
#2d kde graph
sns.displot(data = df, x = 'price', y = 'year', kind = 'kde', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.show()
```

#CDF

```python
sns.displot(data = df, x = 'price', kind = 'ecdf', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.grid(which = 'both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data = df, x = 'price', hue = 'fuelType', kind = 'ecdf', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.grid(which = 'both')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.ecdfplot(data = df, x= 'price', hue = 'fuelType')
plt.grid()
plt.show()
```
