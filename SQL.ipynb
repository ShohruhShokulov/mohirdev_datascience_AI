## Ordering


```python
pd.read_sql("SELECT CustomerID, Country, LastName, Email FROM customers ORDER BY Country",db,index_col = 'CustomerId').head(8)
```

## Creating SQL tables

```python
query = """
CREATE TABLE IF NOT EXISTs worldpop(
  rank INT,
  country TEXT PRIMARY KEY,
  density DOUBLE,
  population DOUBLE NOT NULL,
  area INT NOT NULL
)
"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type ='table'")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(worldpop)")
cursor.fetchall()
```

```python
query = "INSERT INTO worldpop VALUES(1,'Macau',21987,658,30)"
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

```python
query = "INSERT INTO worldpop (country,population,area) VALUES ('Uzbekistan',3600000,448000)"
cursor.execute(query)
db.commit()
```

```python
# updating the info from the DataBase
query = "UPDATE worldpop SET rank = 133, density = 75 WHERE country = 'Uzbekistan'"
cursor.execute(query)
db.commit()
```

```python
pd.read_sql("SELECT * FROM worldpop", db)
```

```python
#Deleting Info from the table
query = "DELETE FROM worldpop WHERE country = 'Macau'"
cursor.execute(query)
db.commit()
```

```python
pd.read_sql("SELECT * FROM worldpop",db)
```

```python
cursor.execute("DROP TABLE worldpop")
db.commit()
```
