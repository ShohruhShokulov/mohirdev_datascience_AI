```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv')
df = df[df.year>2010]
df = df[df.transmission!='Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)

plt.figure(figsize = (12,6))

plt.plot(x, sinx,
         linewidth = 1.5,
         linestyle = '-.',
         color = 'r',
         alpha = 0.7,
         marker = 'h')
plt.title('sin(x) graph')
plt.grid(color = 'tomato', alpha = 0.5, linestyle = '--')

plt.plot(x, cosx,
         linewidth = 1.5,
         linestyle = ':',
         color = 'c',
         alpha = 0.7,
         marker = 'v')
plt.title('cos(x) graph')
plt.grid(color = 'tomato', alpha = 0.5, linestyle = '--')

plt.show()
```

#subplot()

```python
# drawing the graph seperately using subplot() function
x = np.arange(0, 4*np.pi, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)

plt.figure(figsize = (12,6))

plt.subplot(2,1,1)
plt.plot(x, sinx,
         linewidth = 1.5,
         linestyle = '-.',
         color = 'r',
         alpha = 0.7,
         marker = 'h')
plt.title('sin(x) graph')
plt.grid(color = 'tomato', alpha = 0.5, linestyle = '--')

plt.subplot(2,1,2)
plt.plot(x, cosx,
         linewidth = 1.5,
         linestyle = ':',
         color = 'c',
         alpha = 0.7,
         marker = 'v')
plt.title('cos(x) graph')
plt.grid(color = 'tomato', alpha = 0.5, linestyle = '--')

plt.show()
```

```python
sns.histplot(data = df, x = 'price')
plt.show()
```

```python
sns.scatterplot(data = df, x = 'price', y = 'year', s = 100)
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize = (10,10))

#firstplot
sns.histplot(ax = ax[0], data = df, x = 'price')
#secondplot
sns.scatterplot(ax = ax[1],data = df, x = 'price', y = 'year', s = 100)
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize = (15,10))

plt.suptitle('Analysis of the sales of Mercedes cars during 2000 to 2020')
sns.histplot(ax = ax[0,0], data = df, x = 'price')
sns.histplot(ax = ax[0,1], data = df, x = 'mileage')
sns.scatterplot(ax = ax[1,0], data = df, x = 'year', y = 'price', hue = 'fuelType', s = 50)
sns.scatterplot(ax = ax[1,1], data = df, x = 'year', y = 'mileage', hue = 'fuelType', s = 50)

#naming each graph
ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price vs Year correlation")
ax[1,1].set_title("Mileage vs Year correlation")
plt.show()
```
