## dropna()

```python
import pandas as pd
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv",index_col = 'index')
df.head()
```

```python
# checking NaN values in the DataFrame
df.isnull().sum()
```

```python
len(df)
```

```python
#Calculating the proportion of the null values
df.isnull().sum()/len(df)
```

```python
# dropna()method which we can use to drop the row which has NaN value
df.dropna()
```

```python
df.dropna().shape
```

```python
#droping the column which has NaN value
df.dropna(axis = 1)
```

```python
df.dropna(axis = 1,thresh = 48)
```

```python
df.dropna(subset = 'price') # droping the row in which the column of the price is NaN
```

##fillna()

```python
df.fillna(0)
```

```python
df.fillna({'horsepower': 100,'wheel-base':80})
```

```python
df.fillna({'horsepower':df['horsepower'].mean(),'wheel-base':df['wheel-base'].mean()})
```

```python
df.fillna(method = 'ffill')
```

```python
df[df['company'] == 'volkswagen' ].fillna(method = 'bfill')
```

```python
df.fillna(df.mean())
```

```python
df2 = df[df['body-style'] ==  'hatchback']
df2.fillna(df2.median())
```

```python
df2['wheel-base'].mode()[0]
```

```python
df2 = df[df['body-style'] ==  'hatchback']
df2.fillna(df2['wheel-base'].mode()[0])
```

## drop_duplicates()

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/world_population_duplicates.csv")
df.head()
```

```python
df.duplicated().sum()
```

```python
# drop_duplicates method
df.drop_duplicates()
```

```python
df.drop_duplicates(['country','rank'])
```

## map()

```python
# map method
df['num-of-cylinders'].unique()
```

```python
dic =  {'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8}
df['cylinders'] = df['num-of-cylinders'].map(dic)
df
```

```python
# converting the price from USD to UZS and putting it into DataFrame
rate = 12500
df['price_uz']= df['price'].map(lambda x: x*rate)
df
```
