```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numbers = np.arange(10)
numbers
```

```python
plt.plot(numbers)
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
```

```python
plt.figure(figsize = (10,4)) # changing the size of the plot

plt.plot(x,y,linewidth = 2,linestyle = "",color = 'r',marker = "o") #crearing plot
plt.title("sin(x) function") #titling
plt.xlabel('pi') #naming the x axis
plt.ylabel('sin(x)') #naming y axis
plt.grid(color = 'k',linestyle = '--') #changing background
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
sinx = np.sin(x)
cosx = np.cos(x)

#size of the figure
plt.figure(figsize = (12,6))

#plot sinx(x)
plt.plot(x,sinx,
         linewidth = 1.5,
         linestyle = "-.",
         color = 'r',
         alpha = 0.7,
         marker = "h")

#plot cos(x)
plt.plot(x,cosx,
         linewidth = 1.5,
         linestyle = ':',
         alpha = 0.7,
         color = 'c',
         marker = 'v')

#titeling the graph
plt.title("sin(x) and cos(x) functions' graph")

#naming x axis
plt.xlabel('pi')
plt.legend(['sin(x)','cos(x)'],loc = 0, fontsize = "x-large") #putting the names of the graphs on the corner of the graph using legend function
plt.grid(color = 'tomato',alpha = 0.5, linestyle = "--",)

#saving the graph
plt.savefig("graph.png")


#showing the plot
plt.show()
```
