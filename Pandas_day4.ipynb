##   Arithmetic Operations

```python
#Arithmetic methods on pandas
import pandas as pd
import numpy as np
```

```python
#series
s1 = pd.Series([1.2,3.4,1.7,1.4,2.3], index = ['a','b','c','d','e'])
s2 = pd.Series([4.5,6.7,8.9,3.3,4.7], index = ['a','c','f','g','f'])
```

```python
s1
```

```python
s2
```

```python
s1 +s2 # Arithmetic operations are done according to the indexes of the elements
```

```python
# DataFrame
df1 = pd.DataFrame(np.arange(9).reshape(3,3), columns = list('abc'),index = ['day1','day2','day3'])
df2 = pd.DataFrame(np.arange(2,14).reshape(4,3), columns  = list('bcd'), index   = ['day1','day3','day4','day5'])
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
df1.add(df2)
```

```python
df1.add(df2, fill_value= 0)
```

```python
# subtraction
df1.sub(df2, fill_value= 0)
```

```python
df2.div(df1)
```

## Use of Functions on Pandas

```python
df = pd.DataFrame(np.random.randn(3,4), index = list('abc' ), columns = ['Day1','Day2','Day3','Day4'])
df
```

```python
# absolute value
np.abs(df)
```

```python
# rounding
np.round(df)
```

```python
df.mean()
```

```python
df.mean(axis = 1)
```

```python
# apply Function
df.apply(np.round)
```

```python
df.apply(sum)
```

```python
df.apply(sum, axis= 1)
```

```python
# lambda function
f = lambda x: x.max() - x.min()
```

```python
# getting the difference between maximum and minimum of the elements
df.apply(f)
```

```python
def foo(x):
  return pd.Series([x.min(),x.max()], index = ['Min', 'Max'])
```

```python
df.apply(foo)
```

```python
df
```

```python
# applymap Function for every element in the list
df.applymap(lambda x: x*x)
```
