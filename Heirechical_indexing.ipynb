```python
df = pd.read_csv("https://github.com/MansurCompAI/Practicum_Datasets/raw/master/world_cities_air_water_quality.csv", index_col = 0)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
columns = ['Region','Country','Airquality','Waterqality']
df.columns = columns
df.head()
```

```python
df['Region'] = df['Region'].map(lambda x: x.replace('"',""))
df.head()
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"',""))
df.head()
```

```python
df['Country'][1]
```

```python
df['Country'] = df['Country'].map(lambda x: x.strip()) #getting rid of white space from the dataset
df.head()
```

```python
df['Country'][1]
```

```python
df[df.Country == "Germany"]
```

```python
df[df.Country == "Uzbekistan"]
```

## pivot_table()

```python
df1 = pd.pivot_table(df,index = ['Country','City'])
df1
```

```python
df1.loc['Uzbekistan']
```

```python
df1
```

```python
df2 = df1.copy()
df2
```

```python
df2['Airpollution'] = 100 - df2['Airquality']
df2
```

```python
df2['Waterpollution'] = 100 - df2['Waterqality']
df2
```

```python
df3 = pd.DataFrame(df2.values, index = df2.index, columns = [['Air','Air','Water','Water'],['Airquality','Airpollution','Waterqality','Waterpollutin']])
df3.head()
```

```python
df3.columns.names = ["Object","Info"]
df3
```

## swaplevel()

```python
df3.swaplevel('Country','City')
```

```python
# sort_index
df3.sort_index(level= 0, ascending=False)
```
