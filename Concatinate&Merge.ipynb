```python
a = pd.DataFrame({"Transport": ["Train", "Car", "Boat"],
                  "Speed":[150,120,50]})
a
```

```python
b =  pd.DataFrame({"Transport": ["Train", "Car", "plane"],
                  "Price":[15000,1200,500000]})
b
```

```python
merged = pd.merge(a,b)
merged
```

```python
merged_outer = pd.merge(a,b, how = 'outer')
merged_outer
```

```python
merge_left = pd.merge(a,b,how = 'left') #common and all the elements of a
merge_left
```

```python
merge_right = pd.merge(a,b,how = 'right') # common and all the elements of b
merge_right
```

```python
a = pd.DataFrame({"Transport_means": ["Train", "Car", "Boat"],
                  "Speed":[150,120,50]})
a
```

```python
b
```

```python
merge = pd.merge(a,b,left_on = 'Transport_means',right_on = 'Transport' )
```

```python
merge
```

## Merging using indexes

```python
a.set_index('Transport_means',inplace = True)
```

```python
b
```

```python
merged1 = pd.merge(a,b,left_index = True, right_on = "Transport")
merged1
```

```python
b.set_index('Transport', inplace = True)
```

```python
merged2 = pd.merge(a,b,left_index = True, right_index = True)
merged2
```

```python
merged2 = pd.merge(a,b,left_index = True, right_index = True, how = 'outer')
merged2
```
