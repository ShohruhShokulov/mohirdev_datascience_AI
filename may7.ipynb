```python
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
```

```python
plt.figure(figsize = (10,4))

plt.plot(x,y, linewidth  = 2, linestyle  = "", color = "r", marker = "o")
plt.title("sin(x) fucntion")
plt.xlabel("pi")
plt.ylabel("sin(x)")
plt.grid(color = "k", linestyle = "--")
plt.show()
```

```python
x = np.arange(0, 4* np.pi, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)

plt.figure(figsize = (12,6))

plt.plot(x, sinx, linewidth = 1.5,linestyle = "-.", color = "r", alpha = 0.7, marker = "h")
plt.plot(x, cosx, linewidth = 1.5, linestyle = ":", alpha = 0.7, color = "c", marker = "v")

plt.title("sin(x) and cos(x) functions graph")

plt.xlabel("pi")
plt.legend(['sinx', 'cosx'], loc = 0, fontsize = "x-large")
plt.grid(color = "tomato", alpha = 0.5, linestyle = "--")

plt.savefig("graph.png")
plt.show()
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv',index_col = 0,parse_dates = [0],usecols = [0,1])
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).max()
df2.head()
```

```python
df2.unstack(level = 1 )
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(data = df2.unstack(level = 0), cmap = "Blues", linewidths = 1, square = True)
plt.show()
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv')
df = df[df.year >= 2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize = (10,5))
sns.histplot(data = df, x = "price")
plt.show()
```

```python
sns.displot(data = df, x = 'price', height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price', hue = 'transmission', height = 5, aspect = 2 )
plt.show()
```

```python
sns.displot(data = df, x = 'price', hue = 'transmission', multiple = 'stack', height = 5, aspect = 2 )
plt.show()
```

```python
sns.displot(data = df, x = 'price', kind = 'kde', height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price', kde = True, height = 5, aspect = 2)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.kdeplot(data = df, x = 'price', shade = True)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.kdeplot(data = df, x = 'price', hue = 'fuelType', shade = True)
plt.show()
```
