```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/world_population.csv",usecols = ['country','pop2021','area'],index_col = 'country')
df.head()
```

```python
df.loc['Uzbekistan']
```

```python
bins = [0,1000,10000,30000,50000,100000,300000,1500000]
```

```python
population = df.pop2021
```

```python
groups = pd.cut(population,bins)
groups
```

```python
groups.value_counts()
```

```python
df['groups'] = groups
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv",index_col = 0)
df.head()
```

```python
df['price'].describe()
```

```python
bins = [0,10000,20000,50000]
```

```python
range = pd.cut(df.price, bins)
range
```

```python
names = ['cheap','middle','expensive']
range = pd.cut(df.price, bins, labels = names)
range.value_counts()
```

```python
range = pd.cut(df.price, 5)
range.value_counts()
```

```python
# qcut Which splits the DAtaset into same amount of categories
range = pd.qcut(df['horsepower'],3)
range.value_counts()
```

## groupby()

## Cleaning Wrong Data

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/MOCK_USER_DATA.csv',index_col = 'id')
df.head()
```

```python
df.describe()
```

```python
df[(df.age > 123) | (df.age <0)]
```

```python
df['age'] = df['age'].apply(np.abs)
df
```

```python
idx = df[df.age > 123].index
idx
```

```python
df.drop(index = idx,inplace = True)
df.describe()
```

```python
df.describe()
```
